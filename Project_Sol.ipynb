# Python Handling Text Files Project Solution

```python
#Want to use pandas library for read the txt file as csv format
#importing Pandas library

import pandas as pd
```

```python
#Data Eyeballing. 
#Reading the text file as line by line, so using readlines() function.

text_file = open("conv.txt", "r")
text_file.readlines()
```

```python
#Reading the text file in CSV and seperating with columns named Speaker and Speech.
#Using "data" as variable name.
#pd.read_csv is the method to read files in Pandas.
#Used the (sep=":") method to seperate the data and stored it in different columns named Speaker and Speech.

data=pd.read_csv("conv.txt", sep=": ", names=["Speaker", "Speech"])
print(data)
```

```python
# Removing Special Characters from Speech Column.
#After data eyeballing, I found these characters are present inside the file and it's needs to be removed as per the guidelines.
#Used str.replace() method to replace every special characters and used regex method to save after every change.

data["Speech"] = data["Speech"].str.replace('?', '', regex=True)
data["Speech"] = data["Speech"].str.replace('’', '', regex=True)
data["Speech"] = data["Speech"].str.replace('.', '', regex=True)
data["Speech"] = data["Speech"].str.replace(',', '', regex=True)
data["Speech"] = data["Speech"].str.replace('!', '', regex=True)
data["Speech"] = data["Speech"].str.replace('…', '', regex=True)

print(data)
```

```python
#Just eyeballing the data once again by using head() function of Pandas

data.head()
```

```python
#Getting Unique values from Speaker column for creating seperate txt files.
#Using .unique() method of pandas to get uniuqe values in a Speaker Column

print(data["Speaker"].unique())
```

```python
#Setting the Speaker column as a index for dataset for better accessing of data.
#Using the set_index() method of pandas to set the index

data.set_index('Speaker', drop= True, inplace=True)
```

```python
#Just data eyeballing after setting the index of first 5 rows.

data.head(5)
```

```python
#Setting the Speaker column as a index is helpfull in accessing data with the speaker value by using loc[] method of pandas
```

```python
#By using .loc["Speaker value"] will give the seperate datas of that speaker value.

data.loc["ARYA"]
```

```python
#After selecting "ARYA" data sepeartely,
#just changing all the words into lowercase by the method str.lower(). 
#Then spliting all the words by str.split() method.
#Putting all that in a "set" for removing the repeated value
#Sorting all that values for arranging in a alphabetical order.
lists = sorted(set(data.loc['ARYA']['Speech'].str.lower().str.split().sum()))


#Creating a empty list and appending all the values if it is not repeated in the list itself.
f_list = []

for i in lists:
    for p in i:
        if p in f_list:
             i = i.replace(j,"")  
    f_list.append(i)

#Then Writing the list values in the txt file with the respective speaker name.
#Using f, \n, method to print all the words in a new line.
with open('ARYA.txt', 'w', encoding="utf_8",) as q:
    for line in f_list:
        q.write(f"{line}\n")

#Repeated this code for all the all the speaker value in the file.
```
